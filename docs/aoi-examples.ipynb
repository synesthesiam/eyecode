```python
from eyecode import aoi, data, plot
```

# Creating AOIs

```python
code_img = data.busjahn_2013.program_image("basketball")
code_aois = aoi.find_rectangles(code_img)
print code_aois[:3]
```

```python
sub_line_aois = code_aois[code_aois.kind == "sub-line"]
plot.aoi.draw_rectangles(sub_line_aois, code_img)
```

# Hit Testing

```python
raw_fixes = data.busjahn_2013.raw_fixations()
print raw_fixes[:5][["trial_id", "start_ms", "fix_x", "fix_y"]]
```

```python
aoi_fixes = aoi.hit_test(raw_fixes, code_aois)
aoi_cols = aoi.get_aoi_columns(aoi_fixes)
print aoi_fixes[:5][["trial_id", "start_ms", "fix_x", "fix_y"] + aoi_cols]
```

# Utility Methods

```python
padded_aois = aoi.pad(sub_line_aois, 3)
plot.aoi.draw_rectangles(padded_aois, code_img)
```

```python
code_bbox = aoi.envelope(padded_aois)
print code_bbox
```

```python
padded_aois = aoi.add_bbox(padded_aois, code_bbox, "program", "all code")
program_aois = padded_aois[padded_aois.kind == "program"]
plot.aoi.draw_rectangles(program_aois, code_img)
```
